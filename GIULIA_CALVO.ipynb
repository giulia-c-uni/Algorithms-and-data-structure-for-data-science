### Exercise: Selection Sort
Write the function ```SelectionSort(coll)``` that returns a sorted list with the elements in *coll*. 
You have to implements Selection Sort algorithm.

`````python
## Your implementation here!!!

def SelectionSort(coll):
    for i in range(len(coll)-1):           #cominciamo con un for loop fino al penultimo elemento dell'array
        min_pos=i                          #min_pos terrà conto del posto in cui si trova il minimo 
        for j in range (i+1, len(coll)):   #per i valori nell'array che va da i+1 alla fine dell'array
            if coll[min_pos]>coll[j]:      #se il valore nella posizione min_pos è maggiore del valore nella posizione j
                min_pos=j                  #la nuova posione di min sara la pos j
        coll[i], coll[min_pos]= coll[min_pos], coll[i] #switch
    return coll  

print (SelectionSort([7,9,4,6,2,5,1]))
`````

### Exercise: Merge Sort
Complete the implementation of Merge Sort by implementing function ```merge()```.

`````python
## Your implementation here!!!
def merge(A, l, m, r):
    #imposto la lunghezza di due subarray
    n1 = m - l + 1 
    n2 = r - m    
    
    #creo due subarray temporanei
    L = [0] * (n1) 
    R = [0] * (n2)
    
    #inserisco gli ementi di A nei subarray
    for i in range(0 , n1):
        L[i] = A[l + i]     
  
    for j in range(0 , n2): 
        R[j] = A[m + 1 + j] 
  
    # merge i subarrays in A
    i = 0     # indice iniziale del primo subarray
    j = 0     # indice iniziale del secondo subarray
    k = l     # indice iniziale dell'array merged (primo elemento a sinistra dell'array A)
  
    while i < n1 and j < n2 : #finquando ci troviamo dentro i due sub array
        if L[i] <= R[j]:      #se l'elemento che si trova nel subarray sinistro è minore uguale a quello destra
            A[k] = L[i]       #assegnamo il valore nel merged a L[i]
            i += 1            #e aumento il puntatore del sub-array sinistro
        else:                 #se l'elemento che si trova nel subarray destro è minore uguale a quello sinistro
            A[k] = R[j]       #assegnamo il valore nel merged a R[j]
            j += 1            #e aumento il puntatore del sub-array destro
        k += 1                #in entrambi i casi aumento il puntatore dell'array merged 
  
    
    #se siamo fuori dall'array destro (elementi finiti)copia tutti gli elementi dell'array sinistro
    while i < n1:    
        A[k] = L[i] 
        i += 1
        k += 1
  
     #se siamo fuori dall'array sinistro (elementi finiti)copia tutti gli elementi dell'array destro
    while j < n2:     
        A[k] = R[j] 
        j += 1
        k += 1
  
 
def mergeSort_rec(arr,l,r): 
    if l < r:   #se il valore della posizione l è minore del valore dell'ultima posione)
        m = (l+(r-1))//2 #dividi
  
        #sorta in maniera recursiva la parte a sinistra e quella a destra 
        mergeSort_rec(arr, l, m) 
        mergeSort_rec(arr, m+1, r) 
        merge(arr, l, m, r)  #combina i risultati


def mergeSort(B):
    A = B[:]
    n = len(A) # Copy the array just because we decided to return a sorted copy of the original array 
    mergeSort_rec(A, 0, n-1)
    return A
`````

### Exercise: Fractional Knapsack Problem

Your goal: Write a function fractional_knapsack(L,W) which takes a list L of pairs (value, weight) and the capacity  𝑊  and returns maximum possible value we can obtain by selecting items.

`````python
def fractional_knapsack (L,W):
    #salviamo i valori di value e weight in due liste 
    value = [l[0] for l in L] 
    weight = [l[1] for l in L] 
    
    #e indicizziamo
    index = list(range(len(value))) 
    
    #creiamo una zip valore (v) peso (w) e salviamo dentro ratio la divisione tra i due
    ratio = [v / w for v, w in zip(value, weight)] 
    index.sort(key=lambda i: ratio[i], reverse=True) #ordine decrescente dell'index in base al valore ratio 
    
    max_value = 0                #serve a salvare il max value che all'inizio è pari a zero
    fractions = [0] * len(value) #serve a salvare la frazione di ogni oggetto da inserire all'interno
    
    for i in index: 
        if weight[i] <= W:         #se il peso dell'oggetto i è minore del peso che lo zaino può immagazzinare
            fractions[i] = 1       #allora L'INTERO oggetto sarà messo nello zaino
            max_value += value[i]  #aggiorniamo il max value con il valore dell'oggetto
            W -= weight[i]         #e sottraiamo dal peso che lo zaino può portare il peso dell'oggetto
        else:                      # se il peso dell'oggetto i è maggiore del peso che lo zaino può immagazzinare
            fractions[i] = W / weight[i] #immagazzianiamo una frazione di esso 
            max_value += value[i] * W / weight[i] #e aggiungiamo al max value la frazione immagazzinata
            break                  #zaino pieno 

    return max_value
`````

### Exercise: Pareto frontier of a set of points in 2-D space (aka Skyline problem)
We are given a set $S$ of $n$ 2D points.
A point $(x,y)$ dominates a point $(x',y')$ iff $𝑥'\leq 𝑥$ and $y'\leq 𝑦$. 
Our goal is to find the set $P$ of dominating points in $S$. 
This corresponds to find the Pareto frontier (or, equivalently, the skyline). 

The problem can be solved in $\Theta(n\log n)$ time.

To find $P$ we need to sort points in $S$ by $x$ in descending order, 
and if $x$′𝑠 the same by $y$ in descending order. This takes $\Theta(n\log n)$ time. 
Then, we do the following.

- Include first point in $P$ and remember this point as $𝑇$. 
- Iterates through the point (let $C$ current point):
* if $C$ is dominated by $T$, then skip $C$ and go to next point;
* Otherwise, include $C$ in $P$ and set $𝑇=𝐶$.

This step can be performed in linear time.

Implement the function ```pareto_frontier(S)```, which returns the pareto frontier $P$ of the points in $S$.


`````python
## Your implementation here!!!

import functools


def p_cmp (a,b):
    if a[1] != b[1]: return b[1]-a[1] #se la y di a è diversa da quella di b allora ritorna la differna 
                                      #tra la y di b la y di a 
    else: return b[0]- a[0]           #se i due valori di y sono diversi, allora fai la differnza delle x

def pareto_frontier(A):
    A_sorted= sorted(A, key = functools.cmp_to_key(p_cmp) ) 
    b= A_sorted[0]                         #puntatore su A sorted
    D= [A_sorted[0]]                       #lista che contiene il primo elemento di A_sorted
    for i in range (1,len(A_sorted)):      #iteriamo dal secondo elemento di A_sorted
        if A_sorted[i][0] > b[0]:          #a parità di y prenderemo gli elementi con la x maggiore
            D.append(A_sorted[i])          #appendiamo gli elementi con la x maggiore alla lista D
            b= A_sorted[i]                 #resettiamo il contatore
    return (D)                             #lista con i punti appartenenti alla linea di pareto-efficenza



`````

### Exercise: Strongly Connected Components
The goal of this exercise is to implement the following algorithm to compute the strongly connected components of a directed graph $G$. [See here](https://www.hackerearth.com/practice/algorithms/graphs/strongly-connected-components/tutorial/). 

In a directed graph a component is strongly connected if there is a directed path from any vertex to every other vertex of the component. The problem asks to is to partition the graph into maximal strongly connected components.

NetworkX provides a method to compute the strongly connected components of a graph [here](https://networkx.github.io/documentation/stable/reference/algorithms/generated/networkx.algorithms.components.strongly_connected_components.html#networkx.algorithms.components.strongly_connected_components). 

Your goal is to implement the Kosaraju's Linear time algorithm to find Strongly Connected Component. 
The algorithm is described [here](https://www.hackerearth.com/practice/algorithms/graphs/strongly-connected-components/tutorial).

It works in three steps. 

- Do a DFS on the original graph, keeping track of the DFS finish times of each node. This can be done with a stack, when some  finishes put the source vertex on the stack. This way node with highest finishing time will be on top of the stack.
- Reverse the original graph, i.e., if there is an edge $(u,v)$ in the original graph, add the edge $(v,u)$ in the reversed one.
-  Do DFS on the reversed graph, with the source vertex as the vertex on top of the stack. When DFS finishes, all nodes visited will form one Strongly Connected Component. If any more nodes remain unvisited, this means there are more Strongly Connected Component's, so pop vertices from top of the stack until a valid unvisited node is found. This will have the highest finishing time of all currently unvisited nodes.

Take a look at [DFS traversal documentation](https://networkx.github.io/documentation/stable/reference/algorithms/traversal.html#module-networkx.algorithms.traversal.depth_first_search).
Note that the finishing time of a node can be inferred from its position in the DFS tree. 

`````python
## Your implementation here!!!
def strongly_connected_components(DG):
    #primo step
    stack = list(nx.dfs_postorder_nodes(DG)) #tengo traccia del finish time di ogni nodo

    #secondo step
    rev_DG = DG.reverse(copy = True)         #reverse del grafico
    
    #terzo step
    visited=set() 
    scc = []
    
    while stack:                                                   #fino a quando abbiamo elementi dentro la stack
        partenza=stack.pop()                                       #nodo con il tempo finale più alto
        if partenza in visited:
            continue 
            
        else:                                                          #se il nodo non è stato visitato
            preorder_nodes=list(nx.dfs_preorder_nodes(rev_DG,partenza))#usando il reverso e partendo dal nodo 
                                                                       #che stiamo analizzando  
                                                                       #quali nodi posso visitare?
            nuovi_nodi=set()                                           #nodi che posso visitare usando il reverso e 
                                                                       #partendo dal nodo che stiamo analizzando
        
            for nodo in preorder_nodes:                 #aggiungo i nodi che posso visitare nel reverse ai nodi 
                if nodo not in visited:                 #già visitati partendo dal nodo che stiamo analizzando
                    nuovi_nodi.add(nodo)
                    
            visited.update(preorder_nodes)              #aggiorno il set di nodi visitati 
            scc.append(nuovi_nodi)                      #aggiorno gli scc components
        
    return scc

#test sull'implementazione
DG = nx.DiGraph()
DG.add_nodes_from(['0','1','2','3','4','5','6','7'])
DG.add_edges_from([('0','1'),('1','2'),('2','0'),('2','3'),('3','4'),('4','5'),('5','6'),('6','4'),('4','7'),('6','7')])


print("Gli strongly connected components di questo grafo sono",strongly_connected_components(DG))

`````
